```python
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.basemap import Basemap, cm
%matplotlib inline
import numpy as np

url = 'https://raw.githubusercontent.com/Preetikasri/UFO-Sightings-Analysis/master/data/full_final_UFO.csv'
ufo = pd.read_table(url, delimiter= ",", error_bad_lines=False )
fig = plt.figure(figsize=(20,15))
m = Basemap(projection='merc',lat_0=0,lon_0=0, \
			llcrnrlat=26,urcrnrlat= 50,\
			llcrnrlon=-125,urcrnrlon=-40, resolution='h')
			
m.etopo()

m.drawcoastlines()
m.drawstates()


#draw parallels
parallels = np.arange(0.,90,10.)
m.drawparallels(parallels,labels=[1,1,1,1],fontsize=10)
#draw meridians
meridians = np.arange(0.,360.,10.)
m.drawmeridians(meridians,labels=[1,1,1,1],fontsize=10)
x, y =   m(list(ufo['longitude '].astype('double')),list(ufo['latitude'].astype('double')), inverse=False)
m.plot(x, y, "ro", markersize = 2, alpha = 0.5)
plt.show()
```

```python
## Plotting the UFO sightings of the Florida state. We subset the ufo dataset to fl dataset having only observations from FLorida
## python v2.7

fl =  ufo[ufo['state'] == 'fl']
plt.figure(figsize=(12,8))
FL = Basemap(projection='merc', llcrnrlat = 24, urcrnrlat = 31.5, llcrnrlon = -88, urcrnrlon = -80, 
             resolution = 'h')
FL.bluemarble()
#FL.etopo()
FL.drawcoastlines()

par = FL.drawparallels(np.arange(0.,90.,10.),labels=[1,1,1,1])
merid = FL.drawmeridians(np.arange(-90,-70,5),labels=[1,1,1,1])

a, b = FL(list(fl['longitude '].astype(float)), list(fl['latitude'].astype(float)), inverse=False)
FL.plot(a, b, "bo", markersize = 3, alpha = 0.8, color = "yellow")
plt.show()
```

```python
## Plotting the NewYork UFO sightings
## python v2.7
ny = ufo[ufo['state'] == 'ny']
plt.figure(figsize=(12,8))
NY = Basemap(projection='merc', llcrnrlat = 40, urcrnrlat = 45.5, llcrnrlon = -80, urcrnrlon = -71.75, 
             resolution = 'h')
NY.bluemarble()
NY.drawcoastlines()
NY.drawstates()

par = NY.drawparallels(np.arange(35,60,5),labels=[1,1,1,1])
merid = NY.drawmeridians(np.arange(-85,-65,5),labels=[1,1,1,1])

aa, bb = NY(list(ny['longitude '].astype(float)), list(ny['latitude'].astype(float)), inverse=False)
NY.plot(aa, bb, "bo", markersize = 4, alpha = 0.8, color = "yellow")
plt.show()
```
