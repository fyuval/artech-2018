```python
import urllib
import json
import requests
from datetime import datetime
from urllib.request import urlopen
import csv
import time

url='https://data.gov.il/action/package_search?rows=10&start=0'

r=requests.get(url)
print(r)
```

```python
prefix="https://data.gov.il/dataset?q=&sort=title_string+asc&page="

item_list=[]
for i in range(1,31):
    print(i)
    url = ''.join([prefix, str(i)])
    r = requests.get(url)
    sleep(10)
    soup = BeautifulSoup(r.text)
    ul = soup.find_all("h3", {"class":"dataset-heading"})
    for it in ul:
        item_list.append({"link":it.a.get("href"), 
                         "name":it.a.text.strip()})
#content > div.row.wrapper > div > section:nth-child(1) > div.module-content > ul > li:nth-child(9) > div > h3 > a
#heads = [it.find("h3", {"class":"dataset-heading"}) for it in ul]
#print(heads)
```

```python
print(item_list)
```

```python
import pandas as pd
df = pd.DataFrame(item_list)
print(df)
```

```python
df.sort_values(["name"], inplace=True)
print(df)
```
