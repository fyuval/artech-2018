```python
import pandas as pd
from glob import glob
import hashlib
```

```python
def get_sizes():
    gov_dir = "../data/gov_files/"
    gov_pat = ''.join([gov_dir, "*.csv"])
    gov_files = glob(gov_pat)
    res_lst = []
    for f in gov_files:
        print("reading {}".format(f))
        df = pd.read_csv(f)
        print(df.size)
        res_lst.append({"name":f,
                       "num_recs":df.shape[0],
                       "num_vars":df.shape[1]})
    res_df = pd.DataFrame(res_lst)
    print(res_df)
    return res_df
```

```python
sizes_df = get_sizes()
out_path = "../data/inputs/sizes.csv"
sizes_df.to_csv(out_path, index=False)
```

```python
def get_user_code(s):
    return int(hashlib.sha256(s.encode('utf-8')).hexdigest(), 16) % 10**7
```

```python
def get_row_num(user_code, ind):
    print("user_code = {}, ind = {}".format(user_code, ind))
    num_recs = sizes_df.iloc[ind]["num_recs"]
    print("num_recs = {}".format(num_recs))
    return (user_code%num_recs)
```

```python
user_code = get_user_code("here is a nicer string")
print(user_code)
```

```python
def get_row(file_id, row_num):
    df = pd.read_csv(sizes_df.iloc[file_id]["name"], nrows=row_num+1)
    return df.iloc[len(df)-1]
```

```python
for _idx in range(0,19):
    row_num = get_row_num(user_code, _idx)
    row = get_row(_idx, row_num)
    print("============")
    print(row)
    
```
